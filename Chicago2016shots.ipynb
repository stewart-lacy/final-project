```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
nba_shots_Chicago_="shot log CHI.csv"
```

```python
Chicago_shots=pd.read_csv(nba_shots_Chicago_)
df_chicago = pd.DataFrame(Chicago_shots, columns = ["player position","location x","location y","shoot player","quarter","current shot outcome"])
df_chicago
```

```python
df_chicago=df_chicago.rename(columns={"player position":"POSITION","location x": "LOC_X", "location y": "LOC_Y", "shoot player":"PLAYER_NAME","quarter":"QUARTER","current shot outcome":"EVENT_TYPE"})
```

```python
df_chicago
```

```python
df=df_chicago.groupby(["PLAYER_NAME"]).count()
df.sort_values('EVENT_TYPE',ascending=False).head(6)
```

```python
df_chicago_top_player=df_chicago.loc[df_chicago['PLAYER_NAME'].isin
(['Jimmy Butler','Dwyane Wade','Robin Lopez','Nikola Mirotic','Rajon Rondo','Taj Gibson'])]
```

```python
df_chicago_top_player
```
